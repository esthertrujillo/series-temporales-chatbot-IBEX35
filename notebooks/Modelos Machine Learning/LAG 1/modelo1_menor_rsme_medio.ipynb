```python
# Definir fechas de corte personalizadas para algunas empresas
fechas_corte = {
    'PUIG': '2025-01-01',  # Fecha personalizada para PUIG
}

# Fecha de corte predeterminada
fecha_test_default = '2023-04-01'

# Lista de empresas
empresas = ['ACCIONA', 'ACCIONA ENERGÍA', 'ACERINOX', 'AENA', 'AMADEUS', 'ARCELORMITTAL',
            'BANKINTER', 'BBVA', 'CAIXABANK', 'CELLNEX TELECOM', 'ENAGAS', 'ENDESA',
            'FERROVIAL', 'FLUIDRA', 'GRIFOLS', 'IAG', 'IBERDROLA', 'INDITEX', 'INDRA',
            'INM COLONIAL', 'LABORATORIOS FARMA (ROVI)', 'LOGISTA', 'MAPFRE', 'MERLIN PROPERTIES',
            'NATURGY', 'PUIG', 'REE', 'REPSOL', 'SABADELL', 'SACYR', 'SANTANDER', 'SOLARIA ENERGIA',
            'TELEFONICA', 'UNICAJA BANCO']

for empresa in empresas:
    # Obtener el dataframe de la empresa
    # Use the correct name directly instead of the replaced name.
    df_name = f"df_{empresa.replace(' ', '_').replace('(', '').replace(')', '')}"
    df = globals().get(df_name)

    # Check if the dataframe exists
    if df is None:
        print(f"DataFrame not found for {empresa} (name: {df_name})")
        continue # Skip this iteration and proceed to the next company

    # Definir la fecha de corte según la empresa
    fecha_test = fechas_corte.get(empresa, fecha_test_default)

    # Dividir en train y test según la fecha de corte
    train = df[df.index < fecha_test]
    test = df[df.index >= fecha_test]

    # Separar features (X) y target (y) en train
    X_train = train.drop(columns=['Precio_cierre'])
    y_train = train['Precio_cierre']

    # Separar features (X) y target (y) en test
    X_test = test.drop(columns=['Precio_cierre'])
    y_test = test['Precio_cierre']

    # Imprimir tamaño de train y test
    print(f"Conjunto de entrenamiento para {empresa} → Tamaño: {X_train.shape}")
    print(f"Conjunto de test para {empresa} → Tamaño: {X_test.shape}")

```

```python
# Librerías necesarias
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
import numpy as np

# Modelos a probar
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'MLP Neural Network': MLPRegressor(max_iter=500)
}


# Bucle para entrenar y evaluar modelos en cada dataframe
for empresa in empresas:
    # Accedemos a los dataframes correspondientes using globals()
    df = globals().get(f"df_{empresa}")

    # Separar features (X) y target (y) en train
    X_train = train.drop(columns=['Precio_cierre'])
    y_train = train['Precio_cierre']

    # Separar features (X) y target (y) en test
    X_test = test.drop(columns=['Precio_cierre'])
    y_test = test['Precio_cierre']


    print(f"\nEvaluando modelos para {empresa}...")

    # Lista para guardar RMSE de cada modelo
    model_rmse = {}

    # Bucle sobre los modelos
    for model_name, model in models.items():
        # Entrenar el modelo
        model.fit(X_train, y_train) # Use X_train and y_train directly

        # Predecir en el conjunto de test
        y_pred = model.predict(X_test)

        # Calcular el RMSE
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))

        # Guardar el RMSE
        model_rmse[model_name] = rmse

        # Mostrar el RMSE para cada modelo
        print(f"{model_name} - RMSE: {rmse:.4f}")

    # Encontrar el modelo con el menor RMSE
    best_model = min(model_rmse, key=model_rmse.get)
    print(f"Mejor modelo para {empresa}: {best_model} con RMSE: {model_rmse[best_model]:.4f}")
```

```python
# Librerías necesarias
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
import numpy as np

# Modelos a probar
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'MLP Neural Network': MLPRegressor(max_iter=500)
}

# Hiperparámetros para la búsqueda en grid
param_grid = {
    'Linear Regression': {
        'fit_intercept': [True, False]
    },
    'Decision Tree': {
        'max_depth': [5, 10, 15],
        'min_samples_split': [2, 5, 10]
    },
    'Random Forest': {
        'n_estimators': [100, 200],
        'max_depth': [5, 10, 15],
        'min_samples_split': [2, 5, 10]
    },
    'Gradient Boosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7]
    },
    'MLP Neural Network': {
        'hidden_layer_sizes': [(50,), (100,), (150,)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'sgd'],
        'alpha': [0.0001, 0.001]
    }
}

# Diccionario para almacenar los RMSE por cada empresa
empresa_rmse = {empresa: {} for empresa in empresas}

# Bucle para entrenar y evaluar modelos en cada dataframe
for empresa in empresas:
    # Accedemos a los dataframes correspondientes usando globals()
    df = globals().get(f"df_{empresa}")

    # Separar features (X) y target (y) en train
    X_train = train.drop(columns=['Precio_cierre'])
    y_train = train['Precio_cierre']

    # Separar features (X) y target (y) en test
    X_test = test.drop(columns=['Precio_cierre'])
    y_test = test['Precio_cierre']

    print(f"\nEvaluando modelos para {empresa}...")

    # Bucle sobre los modelos
    for model_name, model in models.items():
        # Definir el GridSearchCV
        grid_search = GridSearchCV(model, param_grid[model_name], cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

        # Entrenar el modelo con GridSearch
        grid_search.fit(X_train, y_train)

        # Mejor modelo encontrado por GridSearch
        best_model = grid_search.best_estimator_

        # Predecir en el conjunto de test
        y_pred = best_model.predict(X_test)

        # Calcular el RMSE
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))

        # Guardamos el RMSE en el diccionario
        empresa_rmse[empresa][model_name] = rmse

        # Mostrar el RMSE para cada modelo
        print(f"{model_name} con parámetros {grid_search.best_params_} - RMSE: {rmse:.4f}")

# Ahora calculamos el RMSE medio para cada modelo
model_avg_rmse = {}

# Calcular RMSE medio para cada modelo
for model_name in models.keys():
    # Obtener los RMSE de cada modelo en todas las empresas
    model_rmse_values = [empresa_rmse[empresa][model_name] for empresa in empresas]

    # Calcular el RMSE medio para el modelo
    avg_rmse = np.mean(model_rmse_values)
    model_avg_rmse[model_name] = avg_rmse

# Encontrar el modelo con el menor RMSE medio
best_model = min(model_avg_rmse, key=model_avg_rmse.get)
print(f"\nMejor modelo global: {best_model} con RMSE medio: {model_avg_rmse[best_model]:.4f}")
```
